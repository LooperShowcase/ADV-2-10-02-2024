```python
import numpy as np 
import pandas as pd 
import plotly.express as px

all_data = pd.read_csv("data.csv" , encoding="latin-1")
streams =all_data['streams']
all_data['streams']=pd.to_numeric(streams,errors='coerce')
```

```python
top_streams_by_year =all_data.groupby('released_year')['streams'].idxmax()
most_streams_by_year =all_data.loc[top_streams_by_year]
most_streams_by_year
```
